This dataset captures the details of how CO2 emissions by a vehicle can vary with the different features. This contains data over a period of 7 years.

There are total 7385 rows and 12 columns. There are few abbreviations that has been used to describe the features. I am listing them out here. The same can be found in the Data Description sheet.

Model

    4WD/4X4 = Four-wheel drive

    AWD = All-wheel drive

    FFV = Flexible-fuel vehicle

    SWB = Short wheelbase

    LWB = Long wheelbase

    EWB = Extended wheelbase

Transmission

    A = Automatic

    AM = Automated manual

    AS = Automatic with select shift

    AV = Continuously variable

    M = Manual

    3 - 10 = Number of gears

Fuel type

    X = Regular gasoline

    Z = Premium gasoline

    D = Diesel

    E = Ethanol (E85)

    N = Natural gas

Fuel Consumption

City and highway fuel consumption ratings are shown in litres per 100 kilometres (L/100 km) - the combined rating (55% city, 45% hwy) is shown in L/100 km and in miles per gallon (mpg)

CO2 Emissions

The tailpipe emissions of carbon dioxide (in grams per kilometre) for combined city and highway driving

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X = df[['ENGINESIZE']]
y = df['CO2EMISSIONS']
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size=0.2)
```

### Linear Regression

```python
#Step:1 importing library
from sklearn import linear_model

# Step:2 Object creation 
lin_reg = linear_model.LinearRegression()

# Step:3 fitting traing data
lin_reg.fit (X_train , y_train)

# The coefficients
print ('Coefficients: ', lin_reg.coef_)
print ('Intercept: ',lin_reg.intercept_)
```

### Evaluation Metrics

    1) Mean absolute error (MAE)
    
    2) Residual sum of squares (MSE)

    3) R^2 - Score

```python
from sklearn.metrics import r2_score

# Prediction on training data
y_train_hat = lin_reg.predict(X_train)

# Prediction on test data
y_test_hat = lin_reg.predict(X_test)

print("Mean absolute error: %.2f" % np.mean(np.absolute(y_test_hat - y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_test_hat - y_test) ** 2))
print("Training Accuracy : %.2f" % r2_score(y_train_hat, y_train))
print("Test Accuracy : %.2f" % r2_score(y_test_hat, y_test))
```

### Decision Tree

```python
# Step:1 importing library and object creation
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 5, max_depth = 5)

# Step:2 fitting the training data
dtc.fit(X_train3,y_train3)

# Step:3 Prediction on training data
dtc_train_predict = dtc.predict(X_train3)

# Step:4 Prediction on testing data
dtc_test_predict = dtc.predict(X_test3)
```

#### EVALUATION

#### TRAINING ACCURACY

```python
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,classification_report,confusion_matrix
dtc_train_accuracy = accuracy_score(dtc_train_predict,y_train3)
print("Training accuracy Decision Tree:",dtc_train_accuracy*100)
print("Classification report of training:"'\n',classification_report(dtc_train_predict,y_train3))
```

#### TESTING ACCURACY

```python
dtc_test_accuracy = accuracy_score(dtc_test_predict,y_test3)
print("Testing accuracy Decision Tree:",dtc_test_accuracy*100)
print("Classification report of training:"'\n',classification_report(dtc_test_predict,y_test3))
```
